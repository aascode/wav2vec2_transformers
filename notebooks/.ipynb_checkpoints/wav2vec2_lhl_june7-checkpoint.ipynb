```python
import librosa
import pandas as pd
import os
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
os.chdir("/content/drive/My Drive/Colab Notebooks")
```

```python
#load pre-trained model and tokenizer
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
#load any audio file of your choice
def transcript(x, y):
    speech = {}
    input_values = {}
    logits = {}
    predicted_ids = {}
    transcribe = {}
    for i in range(x, y+1):
        speech[i], rate = librosa.load(
            "lhl_june7/lhl_june7-%d.flac" % i, sr=16000
        )
        input_values[i] = tokenizer(speech[i], return_tensors="pt").input_values
        logits[i] = model(input_values[i]).logits
        predicted_ids[i] = torch.argmax(logits[i], dim=-1)
        transcribe[i] = tokenizer.decode(predicted_ids[i][0])
    return transcribe
```

```python
%%time
transcripts1 = pd.DataFrame.from_dict(transcript(1,5), orient="index").rename(
    columns={0: "Transcribed_Text"}
)

transcripts2 = pd.DataFrame.from_dict(transcript(6,10), orient="index").rename(
    columns={0: "Transcribed_Text"}
)

transcripts3 = pd.DataFrame.from_dict(transcript(11,15), orient="index").rename(
    columns={0: "Transcribed_Text"}
)

transcripts4 = pd.DataFrame.from_dict(transcript(16,20), orient="index").rename(
    columns={0: "Transcribed_Text"}
)

transcripts5 = pd.DataFrame.from_dict(transcript(21,25), orient="index").rename(
    columns={0: "Transcribed_Text"}
)

transcripts6 = pd.DataFrame.from_dict(transcript(26,30), orient="index").rename(
    columns={0: "Transcribed_Text"}
)

transcripts7 = pd.DataFrame.from_dict(transcript(31,35), orient="index").rename(
    columns={0: "Transcribed_Text"}
)

transcripts8 = pd.DataFrame.from_dict(transcript(36,40), orient="index").rename(
    columns={0: "Transcribed_Text"}
)

transcripts9 = pd.DataFrame.from_dict(transcript(41,45), orient="index").rename(
    columns={0: "Transcribed_Text"}
)
```

```python
transcripts = pd.concat(
    [
        transcripts1,
        transcripts2,
        transcripts3,
        transcripts4,
        transcripts5,
        transcripts6,
        transcripts7,
        transcripts8,
        transcripts9,
    ]
)
```

```python
transcripts.shape
```

```python
lhl_june7 = transcripts["Transcribed_Text"].apply(''.join)
```

```python
lhl_june7.to_csv("lhl_june7.txt", sep="\t", index=False)
```
